```python
from bs4 import BeautifulSoup
import requests
```

```python
headers = {
    'Accept': 'text/html, image/avif, image/webp, image/apn, image/svg+xml, */*;q=0.8',
    'Accept-Encoding' : 'gzip, deflate, br',
    'Accept-Lanugage' : 'en-US, en;q-0.9',
    'Connection' : 'keep-alive',
    'Upgrade-Insecure-Requests' : '1',
    'Sec-Fetch-Dest' : 'document',
    'Sec-Fetch-Mode' : 'navigate',
    'Sec-Fetch-User' : '?1',
    'Sec-Fetch-Site' : 'none',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36 OPR/97.0.4719.17'
}
```

```python
# Champion Name Scaper
def champs():
  champion_data_source = requests.get('https://www.leagueoflegends.com/en-ph/champions/?_gl=1*1d0pzw7*_ga*MTAwMTIwNTcyMy4xNjY4NTE2MTkx*_ga_FXBJE5DEDD*MTY4MDc3NDk4NC4xOS4wLjE2ODA3NzQ5ODQuNjAuMC4w&_ga=2.266898952.768884479.1680774986-1001205723.1668516191').text
  soupify_champions = BeautifulSoup(champion_data_source, 'lxml')
  champs = soupify_champions.find_all('a', class_='style__Wrapper-n3ovyt-0 style__ResponsiveWrapper-n3ovyt-4 ehjaZK elVPdk style__Item-sc-13btjky-3 CanXV isVisible isFirstTime')
  for champ in champs:
    champ_name = champ.find('span', class_='style__Text-n3ovyt-3 gMLOLF')
    print(champ_name.text)

champs()
```

```python
def counters(champion):
  champion_counters = requests.get(f'https://u.gg/lol/champions/{champion}/counter', headers=headers).text
  soupify_champions = BeautifulSoup(champion_counters, 'lxml')

  def counters_weakAgainst(champ_name):
    best_win_rate = soupify_champions.find('div', class_='counters-list best-win-rate')
    best_win_rate_champions = best_win_rate.find_all('a', class_='counter-list-card best-win-rate')
    print(f'\n---------------------------Champions STRONG AGAINST {champ_name}------------------------------\n')
    for best_win_rate_champion in best_win_rate_champions:
      champion_name = best_win_rate_champion.find('div', class_='champion-name').text
      champion_winrate = best_win_rate_champion.find('div', class_='win-rate').text
      print(f'{champion_name}    {champion_winrate}')
    print('\n--------------------------------------------------------------------------------\n')

  def counters_strongAgainst(champ_name):
    worst_win_rate = soupify_champions.find('div', class_='counters-list worst-win-rate')
    worst_win_rate_champions = worst_win_rate.find_all('a', class_='counter-list-card worst-win-rate')
    print(f'\n---------------------------Champions WEAK AGAINST {champ_name}------------------------------\n')
    for worst_win_rate_champion in worst_win_rate_champions:
      champion_name = worst_win_rate_champion.find('div', class_='champion-name').text
      champion_winrate = worst_win_rate_champion.find('div', class_='win-rate').text
      print(f'{champion_name}    {champion_winrate}')
    print('\n--------------------------------------------------------------------------------')

  counters_strongAgainst(champion)
  counters_weakAgainst(champion)
```

```python
if __name__ == '__main__':
  champ_name = input("Enter Champion Name: ")
  counters(champ_name.title())
```
